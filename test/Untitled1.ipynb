```python
#Case 1: rough, idiotic plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.set_xlim(-1.0, 1.0)
ax.set_ylim(-1.0, 1.0)
color = 'black'
alpha_max = 0.7
jsteps = 3

colormap = collections.defaultdict(lambda : 'black')
colormap.update({0:'black', 1:'blue', 2:'red', 3:'magenta'})
for i in range(data.shape[0]):
    if i not in [1]:
        continue
    color = colormap[i]
    color = 'black'
    for j in range(0, data.shape[2]-jsteps, jsteps):
        alpha = (j + 1)/data.shape[2]*alpha_max
        ax.plot(data[i, 0, j:j+jsteps+1], data[i, 1, j:j+jsteps+1], color=color, alpha=alpha)

tcircle = np.linspace(0, 2*np.pi, 101)
xcircle = np.cos(tcircle)*0.8
ycircle = np.sin(tcircle)*0.8
ax.plot(xcircle, ycircle, color='black')
```

```python
t = np.linspace(0, 10, 200)
x = np.cos(np.pi * t)
y = np.sin(t)
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

lc = matplotlib.collections.LineCollection(segments, cmap=plt.get_cmap('copper'),
                    norm=plt.Normalize(0, 10))
lc.set_array(t)
lc.set_linewidth(3)

fig2 = plt.figure()
plt.gca().add_collection(lc)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()
```
